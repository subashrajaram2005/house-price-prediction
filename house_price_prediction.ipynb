```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
datasets = pd.read_csv('boston_house_prices.csv')
```

```python
datasets.rename(columns={'MEDV': 'PRICE'}, inplace=True)
```

```python
datasets.head()
```

```python
datasets.isnull().sum()
```

```python
datasets.describe()
```

```python
correlation = datasets.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot = True,annot_kws={'size':8},cmap='Blues')
```

```python
X = datasets.drop(columns=['PRICE'],axis=1)
Y =datasets['PRICE']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
trainpredict = model.predict(X_train)
```

```python
print(trainpredict)
```

```python
s1 = metrics.r2_score(Y_train,trainpredict)
s2 = metrics.mean_absolute_error(Y_train,trainpredict)
```

```python
print('R2 Score:',s1)
print('Mean Absolute Error:',s2)
```

```python
plt.scatter(Y_train,trainpredict)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
testpredict = model.predict(X_test)
sc1 = metrics.r2_score(Y_test,testpredict)
sc2 = metrics.mean_absolute_error(Y_test,testpredict)
print('R2 Score:',sc1)
print('Mean Absolute Error:',sc2)
```

```python
plt.scatter(Y_test,testpredict)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
